```python
# 1. Import library
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report
```

```python
# 3. Load dataset
df = pd.read_csv("/content/Dataset_Diabetes.csv")  # Ganti nama file jika perlu
print("Dataset loaded:")
print(df.head())
```

```python
# 4. Pisahkan fitur dan label
X = df.drop('Outcome', axis=1)  # Fitur input
y = df['Outcome']               # Label output (0 atau 1)
```

```python
# 5. Normalisasi fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 6. Split data (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
# 7. Inisialisasi dan latih model Logistic Regression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# 8. Evaluasi model
y_pred = model.predict(X_test)
print("Akurasi:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
# 9. Input dinamis dari user
print("\nSilakan masukkan data berikut untuk prediksi:")
kolom = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
         'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
user_data = []

for k in kolom:
    val = float(input(f"{k}: "))
    user_data.append(val)
```

```python
# 10. Normalisasi dan prediksi
user_data_scaled = scaler.transform([user_data])
prediction = model.predict(user_data_scaled)
prediction_proba = model.predict_proba(user_data_scaled)
```
